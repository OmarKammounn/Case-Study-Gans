# Details of tomorrow's flights at the chosen airport: departures and arrivals

```python
# Specify the desired date
date_str = "2023-06-01"  # Enter the date in the format "YYYY-MM-DD"

# Convert the date string to datetime object
date_obj = datetime.datetime.strptime(date_str, "%Y-%m-%d")

# Create the start and end time for the date
start_time = date_obj.replace(hour=0, minute=0, second=0).isoformat()
end_time = date_obj.replace(hour=11, minute=59, second=59).isoformat()

# Define the API endpoint URL with the variable date
url = f"https://aerodatabox.p.rapidapi.com/flights/airports/iata/LHR/{start_time}/{end_time}"

# Define query parameters: Optional Parameters provided by AeroDataBox in RapidAPI
params = {
    "withLeg": "true",
    "withCancelled": "true",
    "withCodeshared": "true",
    "withCargo": "true",
    "withPrivate": "true",
    "withLocation": "false"
}

# Define request headers
headers = {
    "X-RapidAPI-Key": "f41dda726emsh342b9a324d66846p12ee46jsnaa4f1545700b",
    "X-RapidAPI-Host": "aerodatabox.p.rapidapi.com",
    "Accept": "application/json"
}

try:
    # Send the GET request to the API
    response = requests.get(url, headers=headers, params=params)
    response_json = response.json()

    # Check if the request was successful
    if response.status_code == 200:
        # Extract departures and arrivals from the response JSON
        departures = response_json.get("departures", [])
        arrivals = response_json.get("arrivals", [])

        # Convert departures and arrivals to dataframes
        departures_df = pd.json_normalize(departures)
        arrivals_df = pd.json_normalize(arrivals)

        # Print the departures dataframe
        print("Departures:")
        print(departures_df)
        print("\n")

        # Print the arrivals dataframe
        print("Arrivals:")
        print(arrivals_df)
    else:
        print(f"Request failed with status code {response.status_code}")
except Exception as e:
    print(f"An error occurred: {e}")
```

# To better understand the steps that I found particularly difficult

#### Define query parameters

In the given code snippet, the step # Define query parameters is used to specify the query parameters that will be included in the API request.

Query parameters are additional parameters appended to the URL of an API request to provide specific instructions or filters for the requested data. In this case, the query parameters are defined using a Python dictionary called params.

Here's the explanation of each query parameter:

- withLeg: This parameter is set to "true" and indicates that the response should include leg information for the flights. A leg represents a specific segment of a flight, such as a takeoff or landing.

- withCancelled: This parameter is set to "true" and indicates that the response should include information about cancelled flights.

- withCodeshared: This parameter is set to "true" and indicates that the response should include information about codeshared flights. A codeshared flight is a flight operated by one airline but marketed and sold by another airline.

- withCargo: This parameter is set to "true" and indicates that the response should include information about cargo flights.

- withPrivate: This parameter is set to "true" and indicates that the response should include information about private flights.

- withLocation: This parameter is set to "false" and indicates that the response should not include location information for the flights.

These parameters allow you to customize the API request and specify the type of flight information you want to retrieve.

#### Extract departures and arrivals from the response JSON

In the given code snippet, the step # Extract departures and arrivals from the response JSON is used to retrieve the "departures" and "arrivals" information from the JSON response obtained from the API.

After making the API request and receiving the response, the response data is stored in the response_json variable.

The get() method is used to extract the values associated with the keys "departures" and "arrivals" from the response_json dictionary.

- departures = response_json.get("departures", []): This line retrieves the value associated with the key "departures" from the response_json dictionary. If the key "departures" exists in the dictionary, its corresponding value is assigned to the departures variable. If the key does not exist, it returns an empty list [].

- arrivals = response_json.get("arrivals", []): This line retrieves the value associated with the key "arrivals" from the response_json dictionary. If the key "arrivals" exists in the dictionary, its corresponding value is assigned to the arrivals variable. If the key does not exist, it returns an empty list [].

By extracting the departures and arrivals separately, we can further process and analyze each set of flight information independently.

To summarize, the departures and arrivals variables are lists containing dictionaries as elements, representing the departure and arrival information, respectively.
